## Overview of this Notebook

In this notebook, we have created visualizations for the website that shows the reader how the variables relate to either math score or test score, the variance of the variable, and the statistical significance of each variable.


```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("poster")
import statsmodels.api as sm
```

```python
mergedf1 = pd.read_csv("mergedf_final.csv")
mergedf =  mergedf1[mergedf1["Funding"] != 0]
```

## School Performance Measured by Math Scores and Survey Scores

Before we do any analysis, we first present a general picture of the distribution of the math scores and survey scores. We can see from the plots below that most schools have a math score of between 660 and 700 and a survey score between 7.5 and 8.

** School Performance Measured by Math Scores **

```python
mean_score = np.mean(mergedf.Mean_Scale_Score)
print mean_score
plt.hist(mergedf.Mean_Scale_Score,color = "green", alpha = 0.4, )
plt.axvline(mean_score, color = "black", label = "mean math score")
plt.xlabel('Math Scores')
plt.ylabel('Count')
plt.title('School Performance Measured by Math Scores')
plt.legend()
```

** School Performance Measured by Survey Scores **

```python
mean_survey = np.mean(mergedf.Aca_s_11)
print mean_survey
plt.hist(mergedf.Aca_s_11, color = "purple", alpha = 0.4)
plt.axvline(mean_survey, color = "black", label = "mean survey score")
plt.xlabel('Survey Scores')
plt.ylabel('Count')
plt.title('School Performance Measured by Survey Scores')
plt.legend()
```

## Relationship between Independent Variables and Math Scores

After doing the regression analysis (from the Predictions, validation, and inference notebook), we found that funding, average class size, female percentage, Balck percentage, and Hispanic percentage are the statistically significant variables and thus the variables that we are really intereted in. Here we present how the relationship between these variables and the dependent variale looks like.

** Log Funding **

```python
from sklearn import linear_model
import statsmodels.formula.api as sm
score_model = sm.ols("Mean_Scale_Score ~ np.log(Funding)", data=mergedf, missing='drop').fit()
plt.scatter(np.log(mergedf.Funding), mergedf.Mean_Scale_Score, alpha = 0.2)
plt.title("Funding against math score")
plt.xlabel("log of Funding")
plt.ylabel("Math Score")
plt.plot(np.log(mergedf.Funding), score_model.params[1] * np.log(mergedf.Funding) + score_model.params[0], color='red')
```

** Average Class Size **

```python
score_model = sm.ols("Mean_Scale_Score ~ Average_Class_Size", data=mergedf, missing='drop').fit()
plt.title("Average Class Size against Mean Math Score")
plt.xlabel("Average Class Size")
plt.ylabel("Mean Math Score")
plt.scatter(mergedf.Average_Class_Size, mergedf.Mean_Scale_Score, alpha = 0.2)
plt.plot(mergedf.Average_Class_Size, score_model.params[1] * mergedf.Average_Class_Size + score_model.params[0], color='red')
```

** Female Percentage **

```python
score_model = sm.ols("Mean_Scale_Score ~ Female_Percentage", data=mergedf, missing='drop').fit()
plt.title("Female Percentage against Mean Math Score")
plt.xlabel("Female Percentage")
plt.ylabel("Mean Math Score")
plt.scatter(mergedf.Female_Percentage, mergedf.Mean_Scale_Score, alpha = 0.2)
plt.plot(mergedf.Female_Percentage, score_model.params[1] * mergedf.Female_Percentage + score_model.params[0], color='red')
```

** Hispanic Percentage **

```python
from sklearn import linear_model
import statsmodels.formula.api as sm
score_model = sm.ols("Mean_Scale_Score ~ Hispanic_Percentage", data=mergedf, missing='drop').fit()
plt.title("Hispanic Percentage against Mean Math Score")
plt.xlabel("Hispanic Percentage")
plt.ylabel("Mean Math Score")
plt.scatter(mergedf.Hispanic_Percentage, mergedf.Mean_Scale_Score, alpha = 0.2)
plt.plot(mergedf.Hispanic_Percentage, score_model.params[1] * mergedf.Hispanic_Percentage + score_model.params[0], color='red')
```

** Black Percentage **

```python
from sklearn import linear_model
import statsmodels.formula.api as sm
score_model = sm.ols("Mean_Scale_Score ~ Black_Percentage", data=mergedf, missing='drop').fit()
plt.title("Black Percentage against Mean Math Score")
plt.xlabel("Black Percentage")
plt.ylabel("Mean Math Score")
plt.scatter(mergedf.Black_Percentage, mergedf.Mean_Scale_Score, alpha = 0.2)
plt.plot(mergedf.Black_Percentage, score_model.params[1] * mergedf.Black_Percentage + score_model.params[0], color='red')
```

## Relationship between Independent Variables and Survey Score (Students' Self Reported Academic Achievement) 

After doing the regression analysis (from the Predictions, validation, and inference notebook), we found that funding, average number of crimes, average class size, and female percentage are all statistically significant variables and thus the variables that we are really intereted in. Here we present how the relationship between these variables and the dependent variables looks like.

```python
survey_model = sm.ols("Aca_s_11 ~ np.log(Funding)", data=mergedf, missing='drop').fit()
plt.scatter(np.log(mergedf.Funding), mergedf.Aca_s_11, alpha = 0.2, color = "orange")
plt.title("Funding against survey score")
plt.xlabel("log of Funding")
plt.ylabel("Survey Score")
plt.plot(np.log(mergedf.Funding), survey_model.params[1] * np.log(mergedf.Funding) + survey_model.params[0], color='blue')
```

```python
survey_model = sm.ols("Aca_s_11 ~ Avg_No_Crim_N", data=mergedf, missing='drop').fit()
plt.scatter(mergedf.Avg_No_Crim_N, mergedf.Aca_s_11, alpha = 0.2, color = "orange")
plt.title("Average Number of Crimes Against Survey Score")
plt.xlabel("Average Number of Crimes")
plt.ylabel("Survey Score")
plt.plot(mergedf["Avg_No_Crim_N"], survey_model.params[1] * mergedf["Avg_No_Crim_N"] + survey_model.params[0], color='blue')
```

```python
survey_model = sm.ols("Aca_s_11 ~ Average_Class_Size", data=mergedf, missing='drop').fit()
plt.scatter(mergedf.Average_Class_Size, mergedf.Aca_s_11, alpha = 0.2, color = "orange")
plt.title("Average Class Size Against Survey Score")
plt.xlabel("Average Class Size")
plt.ylabel("Survey Score")
plt.plot(mergedf["Average_Class_Size"], survey_model.params[1] * mergedf["Average_Class_Size"] + survey_model.params[0], color='blue')
```

```python
survey_model = sm.ols("Aca_s_11 ~ Female_Percentage", data=mergedf, missing='drop').fit()
plt.scatter(mergedf.Female_Percentage, mergedf.Aca_s_11, alpha = 0.2, color = "orange")
plt.title("Female Percentage Against Survey Score")
plt.xlabel("Female Percentage")
plt.ylabel("Survey Score")
plt.plot(mergedf["Female_Percentage"], survey_model.params[1] * mergedf["Female_Percentage"] + survey_model.params[0], color='blue')
```
